## Data Cleaning :-

```python
# Importing panda and numpy
import pandas as pd
import numpy as np
```

```python
# Reading the csv file
df = pd.read_csv(r'application_record.csv')
```

```python
df.drop_duplicates(subset='ID', keep='first', inplace=True)
```

```python
df['CODE_GENDER'] = df['CODE_GENDER'].replace({'M': 'Male', 'F': 'Female'})
```

```python
df.drop(['FLAG_MOBIL', 'FLAG_WORK_PHONE', 'FLAG_PHONE', 'FLAG_EMAIL'], axis=1, inplace=True)
```

```python
df['DAYS_BIRTH'] = df['DAYS_BIRTH'].abs()
df['DAYS_EMPLOYED'] = df['DAYS_EMPLOYED'].abs()
```

```python
df.dropna(inplace=True)
```

## Creating a new column called "AGE" based on the "DAYS_BIRTH" column:-

```python
# Calculate age from DAYS_BIRTH
df['AGE'] = (df['DAYS_BIRTH'] / 365).astype(int)

print(df)
```

## Create a 'LOAN_STATUS' based on Income :-

```python
# Example: Loan approved if income is above a threshold
df['LOAN_STATUS'] = df.apply(lambda row: 'Approved' if row['AMT_INCOME_TOTAL'] > 100000 else 'Denied', axis=1)
```

```python
df.head()
```

## Creating a 'EMPLOYMENT_LENGTH_YEARS' Column based on Days_Employed :-

```python
# Calculate the length of employment in years based on the "DAYS_EMPLOYED" column.

df['EMPLOYMENT_LENGTH_YEARS'] = (df['DAYS_EMPLOYED'] / 365).astype(int)
```

## Creating a 'CREDIT_SCORE' column :-

```python
# Example: Assign scores based on income and employment length
df['CREDIT_SCORE'] = (df['AMT_INCOME_TOTAL'] / 1000) + (df['EMPLOYMENT_LENGTH_YEARS'] * 10)
```

```python
#operating on the column AMT_INCOME_TOTAL and removing the outliers
df['AMT_INCOME_TOTAL'].describe()

#Calculate the IQR
Q1 = df['AMT_INCOME_TOTAL'].quantile(0.25)
Q3 = df['AMT_INCOME_TOTAL'].quantile(0.75)
IQR = Q3 - Q1
print("Q1: ", Q1, "\nQ3: ", Q3, "\nIQR: ", IQR)
#Remove the outliers
print("Shape before removing outliers: ", df.shape)
df = df[~((df['AMT_INCOME_TOTAL'] < (Q1 - 1.5 * IQR)) |(df['AMT_INCOME_TOTAL'] > (Q3 + 1.5 * IQR)))]
print("Shape after removing outliers: ", df.shape)

#operating on the column DAYS_BIRTH and removing the outliers
df['DAYS_BIRTH'].describe()

#Calculate the IQR
Q1 = df['DAYS_BIRTH'].quantile(0.25)
Q3 = df['DAYS_BIRTH'].quantile(0.75)
IQR = Q3 - Q1
print("Q1: ", Q1, "\nQ3: ", Q3, "\nIQR: ", IQR)
#Remove the outliers
print("Shape before removing outliers: ", df.shape)
df = df[~((df['DAYS_BIRTH'] < (Q1 - 1.5 * IQR)) |(df['DAYS_BIRTH'] > (Q3 + 1.5 * IQR)))]
print("Shape after removing outliers: ", df.shape)

#operating on the column DAYS_EMPLOYED and removing the outliers
df['DAYS_EMPLOYED'].describe()

#Calculate the IQR
Q1 = df['DAYS_EMPLOYED'].quantile(0.25)
Q3 = df['DAYS_EMPLOYED'].quantile(0.75)
IQR = Q3 - Q1
print("Q1: ", Q1, "\nQ3: ", Q3, "\nIQR: ", IQR)
#Remove the outliers
print("Shape before removing outliers: ", df.shape)
df = df[~((df['DAYS_EMPLOYED'] < (Q1 - 1.5 * IQR)) |(df['DAYS_EMPLOYED'] > (Q3 + 1.5 * IQR)))]
print("Shape after removing outliers: ", df.shape)
```

# Try to extract new features by grouping existing columns

```python
# Family Size
df['FAMILY_SIZE'] = df['CNT_CHILDREN'] + df['CNT_FAM_MEMBERS']

# Income per Family Member
df['INCOME_PER_FAMILY_MEMBER'] = df['AMT_INCOME_TOTAL'] / df['FAMILY_SIZE']

# Employment Stability
df['EMPLOYMENT_STABILITY'] = df['EMPLOYMENT_LENGTH_YEARS'] / df['AGE']

# Age Group
def categorize_age(age):
    if age < 30:
        return 'Young'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'

df['AGE_GROUP'] = df['AGE'].apply(categorize_age)

# Income Level
def categorize_income(income):
    if income < 50000:
        return 'Low'
    elif income < 100000:
        return 'Medium'
    else:
        return 'High'

df['INCOME_LEVEL'] = df['AMT_INCOME_TOTAL'].apply(categorize_income)

# Family Type
df['FAMILY_TYPE'] = df['NAME_FAMILY_STATUS'].astype(str) + " - " + df['NAME_HOUSING_TYPE'].astype(str)
print(df.head())

```
